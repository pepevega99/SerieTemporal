```python
# Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sktime.utils.plotting import plot_series
```

```python
#Leeme el archivo csv viajeros_diarios
viajeros_diarios = pd.read_csv('viajeros_diarios.csv', sep=',', encoding='latin-1')
# Establece la columna 'fecha' como índice del DataFrame
viajeros_diarios.set_index('fecha', inplace=True)
viajeros_diarios
```

```python
#Lee el archivo csv viajeros_semanales
viajeros_semanales = pd.read_csv('viajeros_semanales.csv', sep=',', encoding='latin-1')
# Establece la columna 'fecha' como índice del DataFrame

viajeros_semanales.set_index('fecha', inplace=True)
viajeros_semanales.head()
```

```python
#Graficame los viajeros diarios, pero bonito, ponme el eje x y el eje y titulos
viajeros_diarios.plot(figsize=(15, 6))
plt.xlabel('Fecha')
plt.ylabel('Viajeros')
plt.show()
```

```python
viajeros_semanales.plot(figsize=(15, 6))
plt.xlabel('Fecha')
plt.ylabel('Viajeros')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd

# Suponiendo que 'viajeros_diarios' es un DataFrame con el índice 'fecha' y la columna 'viajeros'

# Convertir el índice de fecha a tipo datetime si es necesario
viajeros_diarios.index = pd.to_datetime(viajeros_diarios.index)
from sklearn.model_selection import train_test_split

# Dividir los datos, dejando los últimos 30 para el conjunto de prueba
y_train_diario, y_test_diario = train_test_split(viajeros_diarios, test_size=0.2, shuffle=False)

y_train_diario.index = pd.to_datetime(y_train_diario.index)
y_test_diario.index = pd.to_datetime(y_test_diario.index)
# 'train_data' y 'test_data
#Guardame en pickle los datos de train y test
y_train_diario.to_pickle('y_train_diario.pkl')
y_test_diario.to_pickle('y_test_diario.pkl')

#Dime los primeros valores de y_train_diario
y_train_diario.tail()
y_test_diario.head()


```

```python
plot_series(y_train_diario, y_test_diario, labels=["y_train", "y_test"])
```

```python
y_train_semanal, y_test_semanal= train_test_split(viajeros_semanales, test_size=0.2, shuffle=False)
#guardamelo en pickle
y_train_semanal.tail()
y_test_semanal.head()
```

```python
import pandas as pd

y_train_semanal.index = pd.to_datetime(y_train_semanal.index)
y_test_semanal.index = pd.to_datetime(y_test_semanal.index)


plot_series(y_train_semanal, y_test_semanal, labels=["y_train", "y_test"])



```
